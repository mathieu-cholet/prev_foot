```python
import pandas as pd
df_players= pd.read_csv("input/df_players_ligue1.csv",sep=";")
df_matchs= pd.read_csv("input/df_matchs.csv",sep=";")
df_teams=pd.read_csv("input/df_teams.csv",sep=";")
```

```python
df_players.drop(["Unnamed: 0"], axis=1, inplace=True)
df_teams.drop(["Unnamed: 0"], axis=1, inplace=True)
df_matchs.drop(["Unnamed: 0"], axis=1, inplace=True)
```

```python
def func_day(d):
    return d.split()[0]
def func_month(d):
    return d.split()[1]
def func_year(d):
    return d.split()[2]

def win_nul_lost(score):
    if int(score.split("–")[0])>int(score.split("–")[1]):
        return 1 # if score host team is greater than score away team
    elif int(score.split("–")[0])==int(score.split("–")[1]): 
        return 2 # if score host team is equal score away team
    elif int(score.split("–")[0])<int(score.split("–")[1]): 
        return 3 # if score host team is less than score away team

dict_months = {
    "Janvier": "01",
    "Février": "02",
    "Mars": "03",
    "Avril": "04",
    "Mai": "05",
    "Juin": "06",
    "Juillet": "07",
    "Août": "08",
    "Septembre": "09",
    "Octobre": "10",
    "Novembre": "11",
    "Décembre": "12"
}
```

```python
# Traitment on the df_teams dataframe  
df_teams["date"]=df_teams['date'].map(func_day)+'-'+df_teams['date'].map(func_month).map(dict_months)+'-'+df_teams['date'].map(func_year)
df_teams["team"]=df_teams["team"].str[:-8]
df_teams["budget_transfert"]=df_teams["budget_transfert"].str[1:].str.replace(".","")
df_teams[["attaque","milieu","defense","budget_transfert"]] = df_teams[["attaque","milieu","defense","budget_transfert"]].apply(pd.to_numeric)
df_teams["date"]= pd.to_datetime(df_teams["date"],format="%d-%m-%Y" )

# Traitment on the df_players dataframe  
df_players["date"]=df_players['date'].map(func_day)+'-'+df_players['date'].map(func_month).map(dict_months)+'-'+df_players['date'].map(func_year)
df_players["date"]= pd.to_datetime(df_players["date"],format="%d-%m-%Y" )
df_players["taille"]=df_players["taille"].str[:-3] #taille en centimmètre 
df_players["poid"]=df_players["poid"].str[:-3]#poid en kg 
df_players["valeur"]=df_players["valeur"].str[1:].str.replace(".","")#valeur en €
df_players["salaire"]=df_players["salaire"].str[1:].str.replace(".","") #valeur en €
cols_str_to_num=["taille","poid",'valeur','salaire','controles', 'dribbles', 'marquage','tacle_glisse','tacle_debout', 'engagement', 
                 'reactivite', 'placement','intercept', 'vista', 'discipline', 'passe_centre', 'passe_courte','passe_longue',
                 'phy_acceleration', 'phy_endurance', 'phy_force', ' phy_equilibre', 'phy_vitesse', 'phy_agilite', 'phy_detente',
                 'tir_tete', 'tir_frappe', 'tir_finition', 'tir_loin', 'tir_effet','tir_precision', 'tir_penalty', 'tir_volee', 
                 'gardien_placement','gardien_plongeon', 'gardien_main', 'gardien_pied','gardien_reflexe']
df_players[cols_str_to_num] = df_players[cols_str_to_num].apply(pd.to_numeric)
df_players.rename(columns={'club': 'team'}, inplace=True)

# Traitment on the df_matchs dataframe  
df_matchs["result"]=df_matchs["score"].map(win_nul_lost)
```
